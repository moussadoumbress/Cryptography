```python
# La fonction ElGamal_keys doit prodiure une clé publique et une clé secrète à partir
# de la donné d'un nombre premier p.
import random
def generator(p):
    cond=True
    while(cond):
        g=random.randint(2,p-1)
        cond2 = True
        for i in range(2,p-1):
            if pow(g,i,p) == 1:
                cond2 = False
                break
        if(cond2 == True): return g

def ElGamal_keys(p) :
    g=generator(p)
    alpha=random.randint(2,p-2)
    beta=pow(g,alpha,p)
    public_key=(g,beta)
    secret_key=alpha
    return public_key, secret_key
    
print(list(ElGamal_keys(7))[0][0])        
    
```

```python

# La fonction ElGamal_crypt codifie le message m en utilisant la clé publique public_key.
# On peut supposer que m sera un entier entre 0 et p-1 et que la valeur retournée sera
# une paire de nombres.
def ElGamal_crypt(public_key, m) :
    p=public_key[0]
    g=public_key[1]
    beta=public_key[2]
    k=random.randint(2,p-2)
    return pow(g,k,p),(beta**k)*m
print(ElGamal_crypt([7,3,4],17))    

```

```python
def rabin_miller(n, k) :
    if n == 2 or n ==3 :
       return True
    if n % 2 == 0:
       return False
    r, s = 0, n - 1
    while s % 2 == 0:
       r += 1
       s //= 2                     # la partie non paire de n-1 / permet la factorisation en nombre premier
    for _ in range(k):
      a = random.randint(2, n - 1)     
      x = pow(a, s, n)            
      if x == 1 or x == n - 1:
         continue                
      for _ in range(r - 1):      
         x = pow(x, 2, n)        
         if x == n - 1:
            break               
         else:
           return False            
    return True      
```

```python

    
# La fonction ElGamal_decrypt decode le message crypté m  avec la clé publique public_key,
# en utilisant la clé secrète secret_key. On peut supposer que m sera une paire d'entiers entre 0 et p-1.

def ElGamal_decrypt(public_key,secret_key, m) :
    alpha=secret_key
    p=public_key[0]
    return pow(m[0],-alpha,p)*m[1]

```
